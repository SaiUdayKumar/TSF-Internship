# Prediction using Decision Tree Algorithm
## Develping and visualising a desicion tree classifier which would classify any new data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt         #Importing our required libraries
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')            #Reading the data and writing it to our data frame
df.head()
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)         
                                        #splitting the dataset into training and testing set
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dtree = DecisionTreeClassifier()        #instatiating our model
dtree.fit(X_train, y_train)             #fitting our model with the training data
predictions = dtree.predict(X_test)     #predicting the classes
```

```python
from sklearn.metrics import classification_report, confusion_matrix,  accuracy_score
```

```python
print('\n')
print(confusion_matrix(predictions, y_test))
print('\n')
print(classification_report(predictions, y_test))
print('\n')
print('The accuracy of our Model = {}'.format(accuracy_score(predictions, y_test)))
```
